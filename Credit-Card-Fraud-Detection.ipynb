## Splitting the Dataset

```python
# matrix of features
x= df.drop(columns='Class')
# dependent variable
y = df['Class']
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(
     x, y, test_size=0.2, random_state=0)
```

## StandardScaler

```python
sc = StandardScaler()
x_train = sc.fit_transform(X_train)
x_test  = sc.transform(X_test)
```

## Logistic Regression

```python
classifier_lr = LogisticRegression()
classifier_lr.fit(x_train,y_train)
y_pred = classifier_lr.predict(x_test)
acc = accuracy_score(y_test,y_pred)
f1  = f1_score(y_test,y_pred)
prec = precision_score(y_test,y_pred)
rec = recall_score(y_test,y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
results = pd.DataFrame([['Logistic Regression',acc,f1,prec,rec,roc_auc]],
                      columns=['Model','Accuracy','F1','precision','Recall','ROC-AUC-Score'])
results
```

## Random Forest Classifier

```python
classifier_rf = RandomForestClassifier(random_state=0)
classifier_rf.fit(x_train,y_train)
y_pred = classifier_rf.predict(x_test)
acc = accuracy_score(y_test,y_pred)
f1  = f1_score(y_test,y_pred)
prec = precision_score(y_test,y_pred)
rec = recall_score(y_test,y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
results_1 = pd.DataFrame([['Random Forest',acc,f1,prec,rec,roc_auc]],
                      columns=['Model','Accuracy','F1','precision','Recall','ROC-AUC-Score'])
agg_result = results.append(results_1,ignore_index=True)
agg_result
```

## XG Boost Classifier

```python
classifier_xgb = XGBClassifier(random_state=0)
classifier_xgb.fit(x_train,y_train)
y_pred = classifier_xgb.predict(x_test)
acc = accuracy_score(y_test,y_pred)
f1  = f1_score(y_test,y_pred)
prec = precision_score(y_test,y_pred)
rec = recall_score(y_test,y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
results_2 = pd.DataFrame([['XG boost',acc,f1,prec,rec,roc_auc]],
                      columns=['Model','Accuracy','F1','precision','Recall','ROC-AUC-Score'])
agg_result.append(results_2,ignore_index=True)
```

## Dealing with Imbalanced Data

```python
smote = SMOTE(random_state=0)
x_train_resampled,y_train_resampled = smote.fit_resample(x_train,y_train)
```

 (XG_Boost_With_Smote_Oversampling)

```python
xgb_classifier_1 = XGBClassifier(random_state=0)
xgb_classifier_1.fit(x_train_resampled, y_train_resampled)
y_pred = xgb_classifier_1.predict(X_test)
roc_auc = roc_auc_score(y_test, y_pred)

acc = accuracy_score(y_test,y_pred)
f1  = f1_score(y_test,y_pred)
prec = precision_score(y_test,y_pred)
rec = recall_score(y_test,y_pred)
results_2 = pd.DataFrame([['XG boost',acc,f1,prec,rec,roc_auc]],
                      columns=['Model','Accuracy','F1','precision','Recall','ROC-AUC-Score'])

results_2
```
